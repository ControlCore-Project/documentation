```python
import pandas as pd
from google.colab import files

uploaded = files.upload()
filename = list(uploaded.keys())[0]
data = pd.read_excel(filename)
```

```python
print(data)
print(data.columns)
```

```python
read_shared_data_ns = data.loc[3:100,"Shared Memory.1"]
print("Read Shared Memory data(in nsec):")
print(read_shared_data_ns, "\n\n")


read_file_data_ns = data.loc[3:100,"File Sharing.1"]
print("Read File Sharing data (in nsec):")
print(read_file_data_ns)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(read_shared_data_ns, label="Shared Memory")
plt.plot(read_file_data_ns, label="File Sharing")

plt.title("Read method of Shared Memory vs File Sharing Data in Raspberry Pi")
plt.xlabel("Index")
plt.ylabel("Time (nsec)")

plt.ylim(6000, 6000000)

plt.legend()
plt.show()
```

```python
sum_read_shared = read_shared_data_ns.sum()
sum_read_file = read_file_data_ns.sum()

plt.figure(figsize=(10, 6))
plt.bar(["Shared Memory", "File Sharing"], [sum_read_shared, sum_read_file])

plt.title("Sum of Shared Memory and File Sharing Read Data in Raspberry Pi")
plt.ylabel("Sum Time (nsec)")
plt.show()
```

```python
write_shared_data_ns = data.loc[3:100,"Unnamed: 7"]
print("Write Shared Memory data(in nsec):")
print(write_shared_data_ns, "\n\n")


write_file_data_ns = data.loc[3:100,"Unnamed: 10"]
print("Write File Sharing data (in nsec):")
print(write_file_data_ns)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(write_shared_data_ns, label="Shared Memory")
plt.plot(write_file_data_ns, label="File Sharing")

plt.title("Write method of Shared Memory vs File Sharing Data in Raspberry Pi")
plt.xlabel("Index")
plt.ylabel("Time (nsec)")

plt.ylim(30000, 1900000)
plt.legend()
plt.show()
```

```python
sum_write_shared = write_shared_data_ns.sum()
sum_write_file = write_file_data_ns.sum()

plt.figure(figsize=(10, 6))
plt.bar(["Shared Memory", "File Sharing"], [sum_write_shared, sum_write_file])

plt.title("Sum of Shared Memory and File Sharing Write Data in Raspberry")
plt.ylabel("Sum Time (nsec)")
plt.show()
```
